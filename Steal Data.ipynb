```python
import requests
import json
import pandas as pd
import progressbar
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
base_url = 'http://flex.keeneland.com/misc/SearchResults.do?actionName=HorseSearch&paramNames=search_id%5E!%5Esearch_all_mode%5E!%5Esearch_all_string&paramValues=-1%5E!%5EN%5E!%5E'
```

```python
def save_to_df(df, new_result):
    if df is None:
        df = new_result
    else:
        df = pd.concat([df,new_result])
    return(df)

def loop_with_prefix(df, base_url=base_url, headers=headers,  prefix=''):
    pbar = progressbar.ProgressBar()
    chars = 'abcdefghijklmnopqrstuvwxyz'
    for letter in pbar(chars):
        try:
            page = requests.get(base_url + prefix+letter, headers=headers)
            new_result = pd.read_json(page.text)
            df = save_to_df(df, new_result)
        except: 
            print("Error at: " + prefix)
            loop_with_prefix(df, prefix=prefix+letter)
    return(df)

# 
# pbar = progressbar.ProgressBar()
# chars = 'abcdefghijklmnopqrstuvwxyz'
# for letter in pbar(chars):
#     try:
#         page = requests.get(base_url + letter, headers=headers)
#         new_result = pd.read_json(page.text)
#         save_to_df(df, new_result)
#     except:
#         for letter2 in chars:
#             try:
#                 page = requests.get(base_url + letter+letter2, headers=headers)
#                 new_result = pd.read_json(page.text)
#                 save_to_df(df, new_result)
#             except Exception as e:
#                 print(letter+letter2)
#                 print("AHHHHH, WTF, I don't want to do a lot of error handling so just work. It'll be an infinite loop if I recurse.")
#                 print(e.text)
    
```
